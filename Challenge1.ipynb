```python
# Write a program that will accept a date of birth and determine whether one 
# is considered an adult (i.e. 21 years old).
import datetime
tday = datetime.date.today()
year = int(input("year"))
month = int(input("month"))
day = int(input("day"))
bday2 = datetime.date(year,month,day)
days = (tday - bday2).days

def legal(days):
    days = int(days)
    l_age = 21*365
    if l_age >= (days):
        return "illegal"
    else:
        return "legal"
    
print(legal(days))
    

```

```python
# Write a program that will generate a random playing card e.g. '9 Hearts',
# 'Queen Spades' when the return key is pressed.
# Note: Rather than generate a random number from 1 to 52. Create two random 
# numbers â one for the suit and one for the card.
import random
suit = random.randint(1,4)
num = random.randint(1,13)
def w_suit(suit):
    if suit == 1:
        return "Diamond"
    elif suit == 2:
        return "clubs"
    elif suit == 3:
        return "hearts"
    elif suit == 4:
        return "spades"
    else:
        return -1
def w_card(num):
    if num == 11:
        return "Jack"
    elif num == 12:
        return "Queen"
    elif num == "13":
        return "King"
    else:
        return num
print(w_card(num),w_suit(suit))
    
```

```python
# Make a game of rock, paper, scissors against the computer.
# Algorithm
# Tell user to enter either rock,paper or scissors
# Get the response
# Generate a random number from 1 to 3 (1=rock,2=paper,3=scissors)
# Compare user selection and computer selection
# Display who wins.
import random
choice = input("rock/scissors/paper")
com = random.randint(1,3)
def result(choice):
    ans = {'rock':1,'paper':2,'scissors':3}
    if ans[choice] == com:
        return "Draw"
    elif (ans[choice] == 1 and com == 2) or (ans[choice] == 2 and com == 3) or (ans[choice] == 3 and com == 1):
        return "You Lose"
    else:
        return "You Win"

print(result(choice))
```

```python
# For the rock, paper, scissors game above, ensure that the user enters a valid 
# entry. Add a loop structure to play several times and keep a running score.
import random
    
while True:
    choice = input("rock/scissors/paper/end")
    com = random.randint(1,3)
    ans = {'rock':1,'paper':2,'scissors':3}
    if choice == 'end':
        break
    elif choice not in ans:
        print("Choose valid one pls")
    elif ans[choice] == com:
        print("Draw")
    elif (ans[choice] == 1 and com == 2) or (ans[choice] == 2 and com == 3) or (ans[choice] == 3 and com == 1):
        print("You Lose")
    else:
        print("You Win") 
    
```

```python
# Write a program that will give the answer to logic gate questions e.g.
# Enter logic gate : OR
# Enter first input : 1
# Enter second input :0
# Result = 1
# It should work for the logic gates OR, XOR, NAND and NOR gates.
def OR(a,b):
    if a and b == 0:
        return 0
    else:
        return 1
def XOR(a,b):
    if a == b:
        return 0
    else:
        return 1
def NAND(a,b):
    if a and b == 1:
        return 1
    else:
        return 0
def NOR(a,b):
    if a and b == 0:
        return 1
    else:
        return 0
gate = input("Enter Logic gate")
choice1 = input("Enter First input")
choice2 = input("Enter Second input")
list = {'OR':OR,'XOR':XOR,'NAND':NAND,'NOR':NOR}
lgate = list[gate]
result = lgate(choice1,choice2)
print(result)
```

```python
# Write a program that will display all the factors of an input number that
# are bigger than 1.
# (e.g. the factors of the number 12 are 6,4,3 and 2 because they divide
# into 12 exactly).
# Hint: To find out whether a number X is a factor of Y use:
# If Y mod X == 0 (there is nothing remaining when Y is divided by X)
num = input("Enter num:")
def factor(num):
    num = int(num)
    array = []
    for i in range(2,num):
        if num % i == 0:
            array.append(i)
    return array
print(factor(num))
```
